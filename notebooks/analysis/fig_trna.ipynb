```python
import sys
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path

# ── Configuration via paper_data_config ──
sys.path.insert(0, str(Path("..").resolve()))
from paper_data_config import EPISTASIS_PAPER_ROOT

OUT_DIR = EPISTASIS_PAPER_ROOT / "figures"
OUT_DIR.mkdir(parents=True, exist_ok=True)
mm = 1 / 25.4

img_top = mpimg.imread(str(EPISTASIS_PAPER_ROOT / "data" / "trna_figures" / "fig2_embedding_vs_output.png"))
img_bot = mpimg.imread(str(EPISTASIS_PAPER_ROOT / "data" / "trna_figures" / "fig1_three_views.png"))

# Get aspect ratios
ar_top = img_top.shape[1] / img_top.shape[0]  # width/height
ar_bot = img_bot.shape[1] / img_bot.shape[0]

fig_w = 183  # mm, Nature full width

# Top panel: bar charts (wider than tall)
top_h = fig_w / ar_top
# Bottom panel: contact maps (wider than tall, should span full width)
bot_h = fig_w / ar_bot * 0.9

total_h = top_h + bot_h + 5  # 5mm gap

fig = plt.figure(figsize=(fig_w * mm, total_h * mm))

# Position axes manually so both span full width
gap = 5 / total_h  # normalized gap

ax_top = fig.add_axes([0, 1 - top_h/total_h, 1, top_h/total_h])
ax_bot = fig.add_axes([0, 0, 1, bot_h/total_h])

ax_top.imshow(img_top)
ax_top.axis('off')

ax_bot.imshow(img_bot)
ax_bot.axis('off')

for ext in (".png", ".pdf"):
    fig.savefig(OUT_DIR / f"fig_trna_combined{ext}", dpi=600,
                bbox_inches="tight", facecolor="white", edgecolor="none",
                pad_inches=0.02)
plt.show()
print(f"Saved fig_trna_combined to {OUT_DIR}")
```

**Figure 2: Embedding perturbation recovers RNA secondary structure.** **(a)** Pearson correlation of three computational approaches with the tRNA-Arg-TCT-4-1 secondary structure contact map and 3D proximity (1/distance, PDB 1EHZ). Embedding perturbation (blue) measures how mutations at one position change the token-level embedding at every other position. Log-odds (orange) uses the MLM head to measure changes in nucleotide probabilities. Epistasis geometry (green) uses the mean-pooled normalized epistasis residual $R_\text{singles}$. **(b)** Scatter plot of predicted embedding perturbation dependency score versus crystallographic 3D proximity, with known Watson-Crick base pairs highlighted (red diamonds). **(c)** Ground truth secondary structure contact map (top left), embedding perturbation matrix (top right, Pearson r = 0.81 vs contacts), MLM log-odds dependency matrix (bottom left, r = 0.66), and epistasis geometry matrix (bottom right, r = 0.44). The embedding perturbation matrix visually recovers the characteristic cloverleaf pattern with acceptor stem, D-stem, anticodon stem, and TΨC stem clearly resolved.
